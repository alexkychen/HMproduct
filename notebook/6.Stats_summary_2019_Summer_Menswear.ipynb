## Stats summary for 2019 Summer Menswear data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import results csv files
topSell_df = pd.read_csv("../data/MenswearResults/topSell_allTestUsers.csv").iloc[:,1:]
userCF_df = pd.read_csv("../data/MenswearResults/user_based_cf.csv").iloc[:,1:]
userCF_topSell_df = pd.read_csv("../data/MenswearResults/userCF_plus_topSell.csv").iloc[:,1:]
CB_df = pd.read_csv("../data/MenswearResults/content-based.csv").iloc[:,1:]
random_df = pd.read_csv("../data/MenswearResults/random_recommend.csv").iloc[:,1:]
```

## Significance test among and between groups

```python
#Perform the Kruskal-Wallis Test - a non-parametric equivalent of the One-Way ANOVA
from scipy import stats

#perform Kruskal-Wallis Test 
stats.kruskal(CB_df.AOP, userCF_df.AOP, topSell_df.AOP, random_df.AOP)
```

```python
#Perform Dunn's test (pairwise comparison for nonparametric data)
#! pip install scikit-posthocs
import scikit_posthocs as sp

data = [CB_df.AOP, userCF_df.AOP, topSell_df.AOP, random_df.AOP]

sp.posthoc_dunn(data, p_adjust = 'bonferroni')
```

```python
CB_mean = CB_df.AOP.mean()
userCF_mean = userCF_df.AOP.mean()
topSell_mean = topSell_df.AOP.mean()

print("Content-based AOP mean:",round(CB_mean, 5))
print("User-based CF AOP mean:",round(userCF_mean, 5))
print("Top12Sell AOP mean:",round(topSell_mean, 5))
```
