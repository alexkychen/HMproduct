```python
import pandas as pd
```

## 1. transactions_train.csv

```python
%%time
#read transaction.csv
transaction_raw = pd.read_csv("../data/large_data/transactions_train.csv") 
```

```python
transaction_raw.info()
```

```python
transaction_raw.head()
```

```python
#check start date and last date
print("Start date:", transaction_raw.t_dat.min())
print("Last date:", transaction_raw.t_dat.max())
```

```python
#Convert t_dat to datetime 
transaction_raw['t_dat'] = pd.to_datetime(transaction_raw['t_dat'])
```

```python
#add year, month, and week column
transaction_raw['year'] = transaction_raw['t_dat'].dt.isocalendar().year
transaction_raw['month'] = transaction_raw['t_dat'].dt.month
transaction_raw['week'] = transaction_raw['t_dat'].dt.week
```

```python
transaction_raw.head()
```

```python
#count number of transactions for each year
transaction_raw.year.value_counts()
```

```python
#count number of transactions for each month
transaction_raw.month.value_counts()
```

```python
#Extract 2019 Summer (5,6,7) data
mask = (transaction_raw.year == 2019) & (transaction_raw.month >= 5) & (transaction_raw.month <= 7)
transaction_2019_summer = transaction_raw.loc[mask]
transaction_2019_summer
```

```python
#check the week number in 2019 summer data
transaction_2019_summer.week.unique()
```

```python
#count number of transaction each week
transaction_2019_summer.week.value_counts()
```

## 2. articles.csv

```python
%%time

article_raw = pd.read_csv("../data/large_data/articles.csv")
```

```python
article_raw.info()
```

```python
article_raw
```

```python
#merge 2019 summer transactions and article csv
transaction_2019summer_articles = pd.merge(transaction_2019_summer, article_raw, on="article_id", how="left")
```

```python
transaction_2019summer_articles
```

```python
transaction_2019summer_articles.index_group_name.value_counts()
```

```python
customer_buy_group = transaction_2019summer_articles.groupby('customer_id')['index_group_name'].unique()
customer_buy_group
```

## Check among customers who bought Menswear items also bought what other category

```python
index_group_count = {'Menswear':0, 'Ladieswear':0, 'Divided':0, 'Sport':0, 'Baby/Children': 0}
for cid, groups in customer_buy_group.items():
    if 'Menswear' in groups:
        for group_name in list(groups):
            index_group_count[group_name] += 1
index_group_count
```

## Subset Menswear data

```python
mask = transaction_2019summer_articles.index_group_name == 'Menswear'
transaction_2019summer_menswear = transaction_2019summer_articles.loc[mask]
transaction_2019summer_menswear
```

```python
#check weekly amount of transactions
transaction_2019summer_menswear.week.value_counts()
```

```python
## Consine similarity vs. euclidean distance
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity

m = np.array([[0,0.2,0],
              [0,0.8,0]])

print(euclidean_distances(m, m))
print(cosine_similarity(m))
```
